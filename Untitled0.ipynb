```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sbn
```

```
from google.colab import files
uploaded = files.upload()
import io
df = pd.read_csv(io.BytesIO(uploaded['kidney_disease.csv']))
```

```
df.head()
```

```
df[['htn','dm','cad','pe','ane']]=df[['htn','dm','cad','pe','ane']].replace(to_replace={'yes':1,'no':0})
df[['rbc','pc']] = df[['rbc','pc']].replace(to_replace={'abnormal':1,'normal':0})
df[['pcc','ba']] = df[['pcc','ba']].replace(to_replace={'present':1,'notpresent':0})
df[['appet']] = df[['appet']].replace(to_replace={'good':1,'poor':0,'no':np.nan})
df['classification']=df['classification'].replace(to_replace={'ckd':1.0,'ckd\t':1.0,'notckd':0.0,'no':0.0})
df.rename(columns={'classification':'class'},inplace=True)
df.head()
```

```
sbn.countplot(x="class", data=df, palette="hls")
plt.xlabel("Class (1 = disease, 0 = not disease)")
plt.show()
```

```
sbn.countplot(x='age', data=df, palette="mako_r")
plt.xlabel("Age")
plt.show()
```

```
pd.crosstab(df["age"],df["class"]).plot(kind="bar",figsize=(21,6))
plt.title('Chronic Kidney Disease Frequency for Ages (class 0 = Not Disease, class 1 = Disease)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```
plt.scatter(x=df.htn[df["class"]==1], y=df.bp[(df["class"]==1)], c="green")
plt.scatter(x=df.htn[df["class"]==0], y=df.bp[(df["class"]==0)])
plt.legend(["Disease", "Not Disease"])
plt.xlabel("Hypertension")
plt.ylabel("Blood Pressure")
plt.show()
```

```
pd.crosstab(df["appet"],df["class"]).plot(kind="bar",figsize=(10,4),color=['#3498db','#9b59b6' ])
plt.title('Kidney disease based on appetite')
plt.xlabel('Appetite (1 = Good; 0 = Poor)')
plt.legend(["Haven't Disease", "Have Disease"])
plt.ylabel('Frequency')
plt.show()
```

```
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import Imputer
```

```
imputer = Imputer(missing_values = 'NaN', strategy = 'median', axis = 0)
imputer = imputer.fit(P)
P = imputer.transform(P)
df.iloc[:,1:6] = P
df.iloc[:,1:6]
Q =df.iloc[:, 6:10]
print(Q)
imputer = Imputer(missing_values = 'NaN', strategy = 'most_frequent', axis = 0)
imputer = imputer.fit(Q)
Q = imputer.transform(Q)
df.iloc[:,6:10] = Q
df.iloc[:,6:10]
R = df.iloc[:, 10:19]
R
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
imputer = imputer.fit(R)
R = imputer.transform(R)
df.iloc[:,10:19] = R
df.iloc[:, 10:19]
S=df.iloc[:, 19:25]
S
```

```
imputer = Imputer(missing_values = 'NaN', strategy = 'most_frequent', axis = 0)
imputer = imputer.fit(S)
S = imputer.transform(S)
df.iloc[:,19:25] = S
df.iloc[:,19:25]
df.isnull().sum()
df.describe()
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
y=y.astype("int64")
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X= sc.fit_transform(X)
X = sc.transform(X)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
print(y_train)
print(X_train)
y_test
```

```
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```
class_names = [0,1]
fig,ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks,class_names)
plt.yticks(tick_marks,class_names)

#create a heat map
sns.heatmap(pd.DataFrame(cm), annot = True, fmt = 'g')
ax.xaxis.set_label_position('top')
plt.tight_layout()
plt.title('Confusion matrix for k-Nearest Neighbors Model', y = 1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```
